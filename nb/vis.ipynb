```python
from pathlib import Path

import numpy as np
import rasterio.plot
```

```python
def open_ds(fname) -> rasterio.DatasetReader:
    samples = Path("../test_samples")
    return rasterio.open(samples / fname)


def plot(ds: rasterio.DatasetReader):
    mask = ds.read_masks(1).astype(np.float32) / 255
    rasterio.plot.show(ds.read(), cmap='viridis', alpha=mask)
```

```python
merit = open_ds("merit_dem_fixed1_sample.tif")
plot(merit)
```

```python
os50 = open_ds("os_terrain_50_fixed1_sample.tif")
plot(os50)
```
