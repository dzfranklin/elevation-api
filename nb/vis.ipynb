```python
import os
from pathlib import Path

import numpy as np
import rasterio
import rasterio.plot
```

```python
def open_ds(path) -> rasterio.DatasetReader:
    return rasterio.open(os.path.expanduser(path), masked=True)


def open_sample(fname) -> rasterio.DatasetReader:
    samples = Path("../test_samples")
    return rasterio.open(samples / fname, masked=True)


def plot(ds: rasterio.DatasetReader):
    rasterio.plot.show(ds.read(1, masked=True), cmap="viridis")
```

```python
plot(open_sample("cop90_sample.tif"))
```
